**A. Загрузка и первичный обзор**

```python
#импорт библиотек
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#загружаем данные
df=pd.read_csv('google_books_dataset.csv')
```

```python
#вывод первых строк
df.head(5)
```

```python
#вывод последних строк
df.tail(5)
```

```python
#структура данных
df.info()
```

```python
#размер таблицы
print(f'Количество строк, столбцов: {df.shape[0]}, {df.shape[1]}')
```

```python
# проверить типы данных и при необходимости привести их к корректным типам

# переводим published_date в тип datetime
df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce')

# переводим в isbn_13 в тип string предварительно заполнив пропуски на Unknown
df['isbn_13'] = df['isbn_13'].fillna('Unknown').astype(str)

# переводим в page_count в тип integer предварительно заполнив пропуски на 0
df['page_count'] = df['page_count'].fillna(0).astype(int)
df.info()
```

**B. Обработка данных**

Пропущенные значения

```python
#количество пропусков по столбцам
df.isnull().sum()
```

```python
# заполняем пропуски в строковых (категориальных) признаках значением “Unknown”
nun_str_cols = ['title', 'subtitle', 'authors', 'publisher', 'description', 'isbn_10', 'categories', 'currency', 'thumbnail']
for col in nun_str_cols:
  df[col] = df[col].fillna('Unknown')

# заполняем пропуски в published_date фиктивной датой, сохраняя тип datetime
df['published_date'] = df['published_date'].fillna(pd.Timestamp('1900-01-01'))

# заполняем пропуски в average_rating и list_price медианным значением
df['average_rating'] = df['average_rating'].fillna(df['average_rating'].median())
df['list_price'] = df['list_price'].fillna(df['list_price'].median())
```

```python
#количество пропусков после обработки
df.isnull().sum()
```

**C. Визуализация данных**

```python
#гистограмма распределения
plt.figure(figsize=(15, 10)) #размеры

plt.subplot(2, 2, 1)
sns.histplot(df['page_count'], bins=30, kde=True)
plt.title('Распределение количества страниц')

plt.subplot(2, 2, 2)
sns.histplot(df['average_rating'], bins=10, kde=True)
plt.title('Распределение среднего рейтинга')

plt.subplot(2, 2, 3)
sns.histplot(df['ratings_count'], bins=30, kde=True)
plt.title('Распределение количества оценок')

plt.subplot(2, 2, 4)
sns.histplot(df['list_price'], bins=30, kde=True)
plt.title('Распределение цены')

plt.tight_layout()
plt.show()
```

```python
#линейный график публикации книг по годам
# фильтруем, исключая строки с фиктивной датой 1900-01-01, которая использовалась для заполнения пропусков
df_published = df[df['published_date'] != pd.Timestamp('1900-01-01')]
# подсчитываем количество опубликованных книг по годам и сортируем по году
books_per_year = df_published['published_date'].dt.year.value_counts().sort_index()

plt.figure(figsize=(12, 6))
books_per_year.plot(kind='line')
plt.title('Количество опубликованных книг по годам')
plt.xlabel('Год публикации')
plt.ylabel('Количество книг')
plt.grid(True)
plt.show()
```

```python
# Столбчатая диаграмма: Распределение среднего рейтинга
plt.figure(figsize=(10, 6))
sns.barplot(x=df['average_rating'].value_counts().index, y=df['average_rating'].value_counts().values, palette='Greens_r')
plt.title('Распределение среднего рейтинга')
plt.xlabel('Средний рейтинг')
plt.ylabel('Количество книг')
plt.show()
```

```python
# Boxplot: Распределение количества страниц и цены
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['page_count'])
plt.title('Boxplot количества страниц')
plt.ylabel('Количество страниц')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['list_price'])
plt.title('Boxplot цены')
plt.ylabel('Цена')

plt.tight_layout()
plt.show()
```

**D. Группировки и агрегации**

```python
# cреднее значение по группам (средний рейтинг по авторам)
# Выбираем топ-20 авторов, исключая 'Unknown'
top_authors = df[df['authors'] != 'Unknown']['authors'].value_counts().head(20).index

# Группируем по авторам и вычисляем средний рейтинг
avg_rating_by_author = df[df['authors'].isin(top_authors)].groupby('authors')['average_rating'].mean().sort_values(ascending=False)

print('Средний рейтинг по топ-20 авторам:')
print(avg_rating_by_author.head(10))
```

```python
# подсчёт количества книг по издательствам
# Фильтруем, исключая издателей 'Unknown'
df_filtered_publishers = df[df['publisher'] != 'Unknown']

# Группируем по издателю и подсчитываем количество книг в каждой группе, затем берем топ-10
books_by_publisher = df_filtered_publishers.groupby('publisher').size().nlargest(10)

print('Количество книг по топ-10 издательствам:')
print(books_by_publisher)
```

```python
# cуммирование количества страниц по категориям
# Сначала очищаем и нормализуем категории для суммирования
# Если категория - строка и не 'Unknown', берем первую подкатегорию, иначе 'Unknown'
all_categories_cleaned = df['categories'].apply(lambda x: x.split(', ')[0] if isinstance(x, str) and x != 'Unknown' else 'Unknown')

# Исключаем 'Unknown' для суммирования страниц и группируем по очищенным категориям
sum_pages_by_category = df[all_categories_cleaned != 'Unknown'].groupby(all_categories_cleaned)['page_count'].sum().sort_values(ascending=False)

print('Общее количество страниц по категориям (Топ-10):')
print(sum_pages_by_category.head(10))
```
